<h>RANKING EMPLOYEE SALES PERFORMANCE<h>

```python
%%sql 
WITH employee_total AS (
    SELECT e.employee_id, e.first_name, e.last_name, SUM(od.unit_price * od.quantity) as total_sales
    FROM employees e
    JOIN orders o ON e.employee_id = o.employee_id
    JOIN order_details od ON o.order_id = od.order_id
    GROUP BY e.employee_id, e.first_name, e.last_name
)

SELECT * , 
      RANK() OVER(
             ORDER BY total_sales desc) as rank
FROM employee_total
LIMIT 10;
```

<h> 4. RUNNING TOTAL OF MONTHLY SALES <h>

```python
%%sql 

SELECT *
FROM orders o
JOIN order_details od ON o.order_id = od.order_id
LIMIT 5;
```

```python
%%sql 

WITH monthly_sales AS (
    SELECT DATE_TRUNC('month', o.order_date) AS month,
    SUM(od.unit_price * od.quantity) AS total_sales
    FROM orders o
    JOIN order_details od ON o.order_id = od.order_id
    GROUP BY DATE_TRUNC('month', o.order_date)
    ORDER BY month
)

SELECT month, total_sales,
   SUM(total_sales) OVER(ORDER BY month) AS running_total
FROM monthly_sales
ORDER BY month
LIMIT 12;
```

<h> 5. MONTH OVER MONTH SALES GROWTH <h>

```python
%%sql 

WITH monthly_sales AS (
    SELECT DATE_TRUNC('month', o.order_date) AS month,
    SUM(od.unit_price * od.quantity) AS total_sales
    FROM orders o
    JOIN order_details od ON o.order_id = od.order_id
    GROUP BY DATE_TRUNC('month', o.order_date)
    ORDER BY month
),

 Previous_monthly_sales AS (
    SELECT month, total_sales,
    LAG(total_sales) OVER(ORDER BY month) AS Previous_month_sales
    FROM monthly_sales
)

SELECT month, total_sales, Previous_month_sales, 
       (total_sales - Previous_month_sales)/ Previous_month_sales * 100 AS MOM_growth
FROM Previous_monthly_sales
WHERE Previous_month_sales IS NOT NULL
ORDER BY month;
```

```python
%%sql 

WITH customer_order AS (
    SELECT c.customer_id, 
    c.company_name, 
    o.order_id, 
    SUM(od.unit_price * od.quantity) AS order_value
    FROM customers c
    JOIN orders o ON c.customer_id = o.customer_id
    JOIN order_details od ON o.order_id = od.order_id
    GROUP BY c.customer_id, c.company_name, o.order_id
),

avg_order AS (
    SELECT AVG(order_value) AS avg_value
    FROM customer_order
),

category AS (SELECT co.customer_id,
co.company_name, 
co.order_id, 
co.order_value,
CASE 
    WHEN co.order_value > ao.avg_value THEN 'Above average'
     ELSE 'Average/Below average'
END AS order_category
FROM customer_order co , 
     avg_order ao
)

SELECT COUNT(order_category)
FROM category
WHERE order_category = 'Above average';
```

<h> 7. PERCENTAGE SALES OF EACH CATEGORY <h>

```python
%%sql 

WITH category_total_sales AS (
    SELECT c.category_id, 
    c.category_name, 
    SUM(od.unit_price * od.quantity) AS total_sales
    FROM categories c
    JOIN products p ON c.category_id = p.category_id
    JOIN order_details od ON od.product_id = p.category_id
    GROUP BY c.category_id, c.category_name
),

entire_sales AS (
    SELECT SUM(total_sales) AS TOTAL
    FROM category_total_sales
)

SELECT ct.category_id,
       ct.category_name,
       ct.total_sales,
      (ct.total_sales / es.TOTAL * 100) AS percentage_sales
FROM category_total_sales ct
CROSS JOIN entire_sales es
ORDER BY percentage_sales DESC
LIMIT 10;
```

<h> 8 TOP PRODUCTS PER CATEGORY <h>

```python
%%sql 

WITH product_total_sales AS (
    SELECT 
    p.category_id,
    p.product_id, 
    p.product_name, 
    SUM(od.unit_price * od.quantity) AS total_sales
    FROM products p
    JOIN order_details od ON od.product_id = p.product_id
    GROUP BY p.product_id, p.product_name, p.category_id
),

assigned_products AS ( SELECT 
pt.category_id,
pt.product_id, 
pt.product_name, 
pt.total_sales,
row_number() over( partition by pt.category_id
                  order by pt.total_sales DESC) AS row
FROM product_total_sales pt
)

SELECT 
category_id,
product_id, 
product_name, 
total_sales,
row
FROM assigned_products
WHERE row > 3
ORDER BY category_id, row;
```

<h>9. TOP 20% OF CUSTOMERS BY PURCHASE VOLUME <h>

```python
%%sql 

-- Step 1: Calculate the total purchase volume for each customer
WITH customers AS (
    SELECT c.customer_id,
    c.company_name,
    SUM (od.unit_price * od.quantity) AS purchase
    FROM customers c
    JOIN orders o ON c.customer_id = o.customer_id
    JOIN order_details od ON o.order_id = od.order_id
    GROUP BY c.customer_id, c.company_name
),

-- Step 2: rank the customers based on purchase volume
percentile as (
    SELECT cs.customer_id,
    cs.company_name,
    cs.purchase,
    percent_rank() over(order by cs.purchase) as percentiles
    FROM customers cs
)

-- Step 3: Select customers whose rank falls within the top 20%
SELECT customer_id,
company_name,
purchase,
percentiles
FROM percentile
WHERE percentiles > 0.8
ORDER BY percentiles DESC
LIMIT 20;
```

<h> 10. COMPARE EACH EMPLOYEE's SALES PERFORMANCE TO THE AVERGAE SALES PERFORMANCE OF ALL EMPLOYEES TO EVALUATE THEIR CONTRIBUTIONS. <h>

```python
%%sql 

-- Step 1: Calculate the total sales for each employee
WITH employee_sales AS (
    SELECT e.employee_id,
    e.first_name,
    e.last_name,
    SUM (od.unit_price * od.quantity) AS total_sales
    FROM employees e
    JOIN orders o ON e.employee_id = o.employee_id
    JOIN order_details od ON o.order_id = od.order_id
    GROUP BY e.employee_id, e.first_name, e.last_name
),

-- Step 2: calculate the average of the total sales
average_total_sales as (
    SELECT AVG (total_sales) AS avg_sales
    FROM employee_sales
)

-- Step 3: compare each employee sales to the average sales
SELECT es.employee_id,
es.first_name,
es.last_name,
es.total_sales,
av.avg_sales,
CASE 
    WHEN es.total_sales > av.avg_sales THEN 'Above Average'
    ELSE 'Below Average'
END AS performance
FROM employee_sales es, average_total_sales av
ORDER BY es.total_sales desc
LIMIT 20;
```

<h> 11. Analyze the cumulative sales for each product category over a specified period to understand each category's growth. <h>

```python
%%sql 

WITH category_monthly_sales AS (
    SELECT c.category_id, 
    c.category_name, 
    DATE_TRUNC('month', o.order_date) AS month,
    SUM(od.unit_price * od.quantity) AS total_sales
    FROM categories c
    JOIN products p ON c.category_id = p.category_id 
    JOIN order_details od ON p.product_id = od.product_id
    JOIN orders o ON od.order_id = o.order_id
    WHERE o.order_date BETWEEN '2023-01-01' AND '2023-12-31'
    GROUP BY c.category_id, c.category_name, DATE_TRUNC('month', o.order_date)
),

cumulative_sales AS (
    SELECT 
    category_id,
    category_name,
    month,
    SUM(total_sales) OVER (PARTITION BY category_id ORDER BY month) AS cumulative
    FROM category_monthly_sales
)

SELECT 
category_id, 
category_name, 
month,
cumulative
FROM cumulative_sales
ORDER BY category_id, category_name, month
LIMIT 10;

    
```

<h> 12. Identify products that perform exceptionally well in terms of sales. Specifically, products whose total sales exceed the 75th percentile. <h>

```python
%%sql 

WITH product_sales AS (
    SELECT 
    p.product_id,
    p.product_name,
    SUM(od.unit_price * od.quantity) AS total_sales
    FROM products p
    JOIN order_details od ON p.product_id = od.product_id
    GROUP BY p.product_id, p.product_name
),

ranking AS(
    SELECT 
    PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY total_sales) AS percentiles
    FROM product_sales
    
)

SELECT 
ps.product_id,
ps.product_name,
ps.total_sales
FROM product_sales ps
JOIN ranking r ON ps.total_sales > r.percentiles
ORDER BY ps.total_sales DESC;
```
